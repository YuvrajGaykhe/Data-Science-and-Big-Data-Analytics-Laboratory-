### Data Science Concepts Explained

*   **Transformation**: In data preprocessing, transformation refers to applying a mathematical function to a column of data to change its distribution, often to make it more suitable for modeling (e.g., normalizing, standardizing, or converting to a different scale).

*   **MinMaxScaler**: This is a scaling technique that transforms features by scaling each feature to a given range, typically between 0 and 1. It preserves the shape of the original distribution.

*   **StandardScaler**: This scaling technique transforms data so that it has a mean of 0 and a standard deviation of 1 (unit variance). It's useful when features have different scales and the algorithm assumes normally distributed data.

*   **fit_transform()**: This is a method often found in scikit-learn's preprocessing classes (like `MinMaxScaler` or `StandardScaler`). `fit()` calculates the necessary parameters (e.g., mean and standard deviation for StandardScaler) from the training data, and `transform()` applies those parameters to the data to scale it. `fit_transform()` combines these two steps for convenience.

*   **loc & iloc (in Pandas)**:
    *   `loc` is primarily label-based indexing. You use row and column labels to select data. Example: `df.loc[row_label, column_label]`.
    *   `iloc` is primarily integer-position-based indexing. You use integer positions (like array indices) to select data. Example: `df.iloc[row_index, column_index]`.

*   **Skewness**: Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. A distribution can be negatively skewed (tail on the left), positively skewed (tail on the right), or symmetric (zero skewness).
